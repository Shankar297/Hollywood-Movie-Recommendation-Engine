```python
# importing important library
import numpy as np
import pandas as pd
from nltk.stem import WordNetLemmatizer
```

```python
# load dataset
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# merge movies dat and credit data on basis of title
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies = movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# checking missing value
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
# Find duplicate in dataset
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(2)
```

```python
# join two words which having space
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# create tag column using five columns
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
# tags are in list that converted to string
new_df.head(3)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
import nltk
from nltk.corpus import stopwords
sw = stopwords.words()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join([i for i in x.split() if i not in (sw)]))
```

```python
new_df.head(6)
```

```python
lem = WordNetLemmatizer()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join([lem.lemmatize(i) for i in x.split()]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
# find similarity using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
import joblib
joblib.dump(new_df.to_dict(), 'movie_dict.pkl')
```
